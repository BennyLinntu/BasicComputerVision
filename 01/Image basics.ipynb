# Channel transformation, three channels to single channel gray scale map

```python

import cv2 

img = cv2.imread("girl.jpg",cv2.IMREAD_UNCHANGED)

shape = img.shape
print(shape)

if shape[2] == 3 or shape[2] == 4 :
    
    img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    cv2.imshow("gray_image",img_gray)
cv2.imshow("image", img)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

# Channel transformation, single channel to three channel gray scale map

```python

import cv2 

img = cv2.imread("girl_gray.jpg",cv2.IMREAD_UNCHANGED)

shape = img.shape
print(shape)
img_color = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
print(img_color.shape)
cv2.imshow("color_image",img_color)
cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Image three-channel separation

```python

import cv2
src=cv2.imread('split.jpg')
cv2.imshow('before',src)

b,g,r=cv2.split(src)

cv2.imshow('blue',b)
cv2.imshow('green',g)
cv2.imshow('red',r)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

import numpy as np
import cv2            
 
image=cv2.imread("split.jpg")
cv2.imshow("src",image)
cv2.waitKey(0)
B,G,R = cv2.split(image)

zeros = np.zeros(image.shape[:2],dtype="uint8")
cv2.imshow("BLUE",cv2.merge([B,zeros,zeros]))
cv2.imshow("GREEN",cv2.merge([zeros,G,zeros]))
cv2.imshow("RED",cv2.merge([zeros,zeros,R]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Image channel merging


```python

import cv2
src=cv2.imread('split.jpg')
cv2.imshow('before',src)

b,g,r=cv2.split(src)

g[:] = 0

img_merge=cv2.merge([b,g,r])
cv2.imshow('merge',img_merge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Summary of channel separation and merger

```python
import cv2 as cv
import numpy as np
src=cv.imread('split.jpg')

cv.imshow('before',src)
cv.waitKey(0)

b,g,r=cv.split(src)
cv.imshow('blue',b)
cv.imshow('green',g)
cv.imshow('red',r)
cv.waitKey(0)

src=cv.merge([b,g,r])
cv.imshow('merge',src)
cv.waitKey(0)

src[:,:,2]=100
cv.imshow('single',src)
cv.waitKey(0)
cv.destroyAllWindows()
```

# RGB and BGR conversion

```python
import cv2 
import matplotlib.pyplot as plt 
img = cv2.imread("test2.png", cv2.IMREAD_COLOR)
cv2.imshow("Opencv_win", img)

img_cv_method = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_numpy_method = img[:,:,::-1] 
plt.subplot(1,3,1)
plt.imshow(img_cv_method)
plt.subplot(1,3,2)
plt.imshow(img_numpy_method)
plt.subplot(1,3,3)
plt.imshow(img)
plt.savefig("./plt.png")
plt.show()
cv2.imwrite("opencv.png", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# RGB and HSV conversion

```python
import cv2

def color_space_demo(image):
    gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    cv2.imshow('gray',gray)
    hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    
    cv2.imshow('hsv',hsv)

src=cv2.imread('girl.jpg')
cv2.imshow('before',src)

color_space_demo(src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Histogram drawing method 1

```python
from matplotlib import pyplot as plt
import cv2

girl = cv2.imread("girl.jpg")
cv2.imshow("girl", girl)

plt.hist(girl.ravel(), 256, [0, 256])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Histogram drawing method 2

```python
from matplotlib import pyplot as plt
import cv2
import numpy as np
img = cv2.imread('girl.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap=plt.cm.gray)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])


plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

# Three channel histogram drawing

```python
from matplotlib import pyplot as plt
import cv2
girl = cv2.imread("girl.jpg")
cv2.imshow("girl", girl)
color = ("b", "g", "r")
#Use the for loop to iterate through the list of colors, and the enumerate enumeration returns the index and value
for i, color in enumerate(color):
    hist = cv2.calcHist([girl], [i], None, [256], [0, 256])
    plt.title("girl")
    plt.xlabel("Bins")
    plt.ylabel("num of perlex")
    plt.plot(hist, color = color)
    plt.xlim([0, 260])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
